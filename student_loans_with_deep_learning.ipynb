# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display a sample of y
y[0:5]
```

```python
# Define features set X by selecting all columns but credit_ranking
X= loans_df.drop(columns="credit_ranking")

# Review the features DataFrame
X.head()
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
# Assign the function a random_state equal to 1
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
scaler.fit(X_train)

# Transform/scale training and test X data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
feature_count = len(X.columns)
display(X.columns)

# Review the number of features
print("number of features: " , feature_count)
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer_1 = 8

# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer_2 = 5
# Review the number hidden nodes in the first layer
print(f"The number of hidden nodes in the first layer is: {hidden_nodes_layer_1}")
# Review the number hidden nodes in the second layer
print(f"The number of hidden nodes in the second layer is: {hidden_nodes_layer_2}")

# Define the number of neurons in the output layer
output_neurons = 1
```

```python
# Create the Sequential model instance
nn_model = Sequential()

# Add the first hidden layer
# Add the second hidden layer
# Add the output layer to the model specifying the number of output neurons and activation function
nn_model.add(Dense(units=hidden_nodes_layer_1, activation="relu", input_dim = feature_count))
nn_model.add(Dense(units=hidden_nodes_layer_2, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Display the Sequential model summary
nn_model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 50 epochs and the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=200) #perform, validation_split=0.2
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
# Display the model loss and accuracy results
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Student Loan Recommendation System

## 1. Data Required for a Student Loan Recommendation System

To build an effective recommendation system for student loans, the following data would be essential:

### a) **Student Demographics and Financial Data**:
- **Income**: Determines eligibility for need-based financial aid.
- **Credit Score**: Relevant for private loans to assess risk and determine interest rates.
- **Current Employment/Income Prospects**: Helps determine loan repayment ability.
- **Marital and Family Status**: If the student has dependents, their financial needs may differ.

### b) **Educational Information**:
- **Institution and Degree Program**: Loan types or scholarships might be available depending on whether the student is at a private, public, or technical school.
- **Field of Study**: Some fields (e.g., STEM) may lead to higher-earning jobs, making students eligible for different loan offers.
- **Academic Performance**: High-performing students may qualify for merit-based loans or scholarships.

### c) **Loan Information**:
- **Loan Types**: Federal loans, private loans, subsidized vs. unsubsidized loans.
- **Interest Rates**: Important for evaluating affordability.
- **Repayment Terms**: Duration, deferment periods, income-based repayment options.
- **Loan Amount**: How much is needed for tuition, living expenses, etc.

### d) **Repayment History**:
- **Loan Default Data**: Important for evaluating future risks for specific demographic profiles.
- **Success Stories**: Graduates who took certain loans and were successful in repayment or gaining employment could help in recommending similar loans.

### Relevance and Appropriateness:
This data is relevant because it focuses on factors that directly influence student loan eligibility, repayment capacity, and alignment with a student’s academic and financial goals. Collecting this information helps create personalized loan recommendations that suit the individual student's needs.

---

## 2. Filtering Method

Based on the data, the recommendation system would likely use **content-based filtering**. Here’s why:

### a) **Content-Based Filtering**:
This approach matches students to loan products based on their attributes (income, academic background, etc.) and the loan product’s characteristics (interest rates, repayment options, etc.). The system would evaluate the suitability of different loan options by matching the student's financial and educational profile to loan requirements.

- **Why content-based filtering**: The attributes of both the student and the loan product are clearly defined. For instance, a student with a high credit score and stable employment would be recommended loans with lower interest rates. Conversely, students with lower incomes might receive recommendations for federal loans or loans with flexible repayment terms.

### b) **Collaborative Filtering**:
While collaborative filtering uses the preferences or behaviors of similar users to make recommendations, it is less applicable here since loans are highly personalized. Just because one student chose a specific loan does not necessarily mean it would be suitable for another student with a different financial situation.

### c) **Context-Based Filtering**:
This approach could be useful if you include external data like the student's geographic location (which can influence loan options) or job prospects post-graduation. However, the core recommendation system would still be content-based since it primarily relies on matching attributes of students and loans.

---

## 3. Real-World Challenges

### a) **Data Privacy and Security**:
Handling sensitive financial, academic, and personal data is a significant concern. Ensuring that all data is securely stored and managed in compliance with regulations like **FERPA** (Family Educational Rights and Privacy Act) or **GDPR** is critical to protect students' privacy.

- **Why it’s a concern**: A data breach could expose students' sensitive financial information, which could lead to identity theft or misuse. This could also damage the trust in the recommendation system, reducing adoption.

### b) **Fairness and Bias**:
A recommendation system must avoid introducing biases, especially those that favor certain demographic groups over others. For example, students from wealthy backgrounds or those pursuing higher-paying fields like STEM could be given disproportionately favorable loan options, disadvantaging others.

- **Why it’s a concern**: If the model is not carefully designed, it could unintentionally widen educational and financial inequality. For instance, students with lower incomes or in less lucrative fields may struggle to find affordable loan options, limiting their access to higher education.
